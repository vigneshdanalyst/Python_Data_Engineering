```python
n=5 


for i in range(1, n+1):
    print(" " *(n-i) + "* " * i)

for i in range (n-1,0,-1):
    print(" " * (n-i) + "* " * i)
```

```python
n = 5

# Top half
for i in range(1, n + 1):
    print(" " * (n - i) + "* " * i)

# Bottom half
for i in range(n - 1, 0, -1):
    print(" " * (n - i) + "* " * i)
```
