```python
#Q1. Create a list called tb_cases with these values:
#[120, 85, 200, 45, 310, 95, 178, 60]
#Print the first, last, and middle two values (index 3 and 4).

tb_cases = [120,85,200,45,310,95,178,60]

print(tb_cases[0])
print(tb_cases[-1])
print(tb_cases[3:5])


#Q2. From the same list, print:

#Total cases
#Highest cases
#Lowest cases
#Average cases (hint: sum divided by len)


total_cases= sum(tb_cases)
highest_cases= max(tb_cases)
lowest_cases= min(tb_cases)
average_cases= total_cases / len(tb_cases)

print(total_cases)
print(highest_cases)
print(lowest_cases)
print(average_cases)


#Add 250 to the list, sort it, and print the final list.

tb_cases.append(250)
tb_cases.sort()
print(tb_cases)


#Q4. What will this print?
districts = ["Chennai", "Madurai", "Coimbatore", "Salem"]
print(districts[1:3])

"Madurai","Coimbatore"

score = [88,92,75,60,95,83]

for s in score:
    if s>80:
        print(s)
```
